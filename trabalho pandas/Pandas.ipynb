```python
not_nan_imdb = titles_df[['imdb_score', 'title']].dropna()
display(not_nan_imdb.max())
display(not_nan_imdb.min())
```

```python
only_films = titles_df.loc[titles_df['type'] == 'MOVIE']
soma_tempo_filmes = only_films['runtime'].sum()
media_duracao_filmes = soma_tempo_filmes/only_films['runtime'].count()
print("meida tempo filmes:", media_duracao_filmes)

only_shows = titles_df.loc[titles_df['type'] == 'SHOW']
soma_tempo_shows = only_shows['runtime'].sum()
media_duracao_shows = soma_tempo_shows/only_shows['runtime'].count()
print("meida tempo shows:", media_duracao_shows)
```

```python
only_films = titles_df.loc[titles_df['type'] == 'MOVIE']
print('filme mais longo: ',only_films['runtime'].max())

only_shows = titles_df.loc[titles_df['type'] == 'SHOW']
print('show mais longo: ',only_shows['runtime'].max())
```

```python
only_films = titles_df.loc[titles_df['type'] == 'MOVIE']
not_nan_imdb = only_films[['imdb_score', 'release_year']].dropna()
ano_filmes_cont = not_nan_imdb['release_year'].value_counts()
display(ano_filmes_cont.head(1))
```

```python
import matplotlib.pyplot as pp

only_films = titles_df[titles_df['type'] == 'MOVIE']
media_scores_por_ano_filmes = only_films.groupby('release_year')['imdb_score'].mean().reset_index()
media_scores_por_ano_filmes.columns = ['release_year', 'media_imdb_score']
scatter = pp.scatter(media_scores_por_ano_filmes['media_imdb_score'], media_scores_por_ano_filmes['release_year'], c=media_scores_por_ano_filmes['release_year'], cmap='plasma', alpha=0.7)
pp.colorbar(scatter, label='Release Year')

pp.xlabel('IMDb Score')
pp.ylabel('Release Year')
pp.title('IMDb Score X Release Year por filme')

pp.show()

only_shows = titles_df[titles_df['type'] == 'SHOW']
media_scores_por_ano_shows = only_shows.groupby('release_year')['imdb_score'].mean().reset_index()
media_scores_por_ano_shows.columns = ['release_year', 'media_imdb_score']
scatter = pp.scatter(media_scores_por_ano_shows['media_imdb_score'], media_scores_por_ano_shows['release_year'], c=media_scores_por_ano_shows['release_year'], cmap='plasma', alpha=0.7)
pp.colorbar(scatter, label='Release Year')

pp.xlabel('IMDb Score')
pp.ylabel('Release Year')
pp.title('IMDb Score X Release Year por show')

pp.show()
```
